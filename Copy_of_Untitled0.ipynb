```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://query.data.world/s/kqtk56fwakvdrre6izb7vmlkcmexsz')

# Changed 'NaN' values to blank
NaN_to_blank_df = df.replace(np.NaN,'',regex=True)

print(NaN_to_blank_df.to_string())
```

```python
# Since I was not able to find any visible errors in the data, I decided to replace blanks back to 'NaN'
Blank_to_NaN_df = NaN_to_blank_df.replace(r'^\s*$', np.NaN, regex=True)

print(Blank_to_NaN_df.to_string())
```
